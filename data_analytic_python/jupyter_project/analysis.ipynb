```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/CunyLaGuardiaDataAnalytics/datasets/master/faithful.csv', index_col=0)
```

```python
df.head()#show the first 5 rows of the dataframe
df.tail()#show the first 5 rows of the dataframe
df.head(10)#show the first 10 rows of the dataframe
```

```python
#analyzing the data
# df.hist()
ax = df.hist(bins=15, alpha=0.85, figsize=(12, 5), grid=False)

plt.suptitle("Distribution of Eruptions and Waiting Times",
             fontsize=18, fontweight='bold', y=1.05)

for a, col in zip(ax.flatten(), df.select_dtypes('number')):
    a.set(xlabel=col.capitalize(), ylabel="Frequency", facecolor="#f7f9fa")
    a.tick_params(labelsize=11)

plt.tight_layout()
plt.show()


```

```python
# How many non-null datapoints are in each column?
df.count()
# Finding range of values (max and min)
df.max()
df.min()
# Mean (average)
df.mean()
# Median (middle)
df.median()
# All at once: summary statistics
df.describe()

```
